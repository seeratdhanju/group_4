```python
new_df
data=new_df.groupby(["batsman"],as_index=False)["total_runs"].sum().sort_values(by="total_runs",ascending=False)
data=data.head(10)

data
```

```python
plt.plot(data["batsman"],data["total_runs"],color="blue",linewidth=2)
plt.title("top 10 batsman runs")
plt.xlabel("batsman")
plt.ylabel("total_runs")
plt.xticks(rotation="vertical")
plt.figure(figsize=(5,5))
```

```python
df_a=new_df.groupby(["bowler"],as_index=False)["wide_runs"].sum()
df_a=df_a[df_a["wide_runs"]>100]
df_a
```

```python
a=["red","black","yellow","pink","purple","green","blue"]
plt.bar(df_a["bowler"],df_a["wide_runs"],color=a)
plt.title("bowler performance")   
plt.xlabel("bowler")
plt.ylabel("wide_runs")
plt.figure(figsize=(5,6))
                                                 
```
